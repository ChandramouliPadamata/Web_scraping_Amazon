### Import libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
#getting blocked by Target server somtimes. So, we need to use User-Agent to not blocked from the sites

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.59'}
```

### Get the request from Amazon

```python
url = 'https://www.amazon.in/s?k=watches'
page= requests.get(url, headers = headers)
print(page)             #HTTP responce status code is [200] which is successfull responce 
```

### Parse HTML page

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup.prettify()      # fOR BETTER content
```

### Get the Product Description

```python
names = soup.find_all('a', class_='a-link-normal s-link-style a-text-normal')
names
```

```python
# Below code will generate the only PRODUCT DESCRIPTION.

Product_Description= []
for i in range(0,55):
    Product_Description.append(names[i].get_text())
Product_Description
```

### Get the Brand

```python
brand = soup.find_all('span', class_='a-size-base-plus a-color-base')
brand
```

```python
# Below code will generate the only PRODUCT BRAND.

Brand= []
for i in range(0,55):
    Brand.append(brand[i].get_text())
Brand
```

### Ratings of the Product

```python
ratings = soup.find_all('span', class_='a-icon-alt')
ratings
```

```python
# Below code will generate the only PRODUCT RATINGS.

Ratings= []

for i in range(0,55):
    Ratings.append(ratings[i].get_text())
Ratings
```

### Get the price of the Product

```python
discount_price = soup.find_all('span',class_='a-price-whole')
discount_price
```

```python
# Below code will generate the only PRODUCT PRICE.

Discount_price= []

for i in range(0,55):
    Discount_price.append(discount_price[i].get_text())
Discount_price
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame()      #Create empty Dataset          
```

```python
df['Product_Description']= Product_Description
df['Brand']= Brand
df['Rating_of_the_product']= Ratings
df['Product_Price']= Discount_price
df
```
